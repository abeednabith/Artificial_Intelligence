```python
import pandas as pd

df = pd.read_csv("Q2_GINI_DT.csv")
df.head()
```

```python
# drop the profit coloumn from original dataset and set to target
inputs = df.drop("Profit", axis='columns')
target = df["Profit"]
inputs
```

```python
# encode the label values to corresponding numerical values using LabelEncoder package
from sklearn.preprocessing import LabelEncoder
age = LabelEncoder()
competition = LabelEncoder()
type = LabelEncoder()

inputs['Age_n'] = age.fit_transform(inputs['Age'])
inputs['Competition_n'] = competition.fit_transform(inputs['Competition'])
inputs['Type_n'] = type.fit_transform(inputs['Type'])

# create new input dataframe with corresponding label values
inputs_n = inputs.drop(['Age', 'Competition', 'Type'], axis='columns')
inputs
```

```python
# create Decision tree classifier from sklearn package
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(inputs_n.values,target)
model.score(inputs_n.values,target)
```

```python
# we can predict for Age = Mid, Competition = No, Type = Hardware
print(f"The predicted profit is : {model.predict([[0,0,0]])}")

# test case from original dataset
print(f"The predicted profit is : {model.predict([[2,0,0]])}")
```
